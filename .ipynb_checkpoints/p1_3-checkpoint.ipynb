```python
import numpy as np
import os
import matplotlib.pyplot as plt
from textwrap import wrap
from adjustText import adjust_text
import math

fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlabel('\n'.join(wrap("Years since 2010", 40)))
ax.set_ylabel('\n'.join(wrap("New Housing Units in Albuquerque", 40)))

# Logistic Model 1
line_m1 = np.linspace(0, 64, 100)
polyspace_m1 = ((1.02016/1.02493)**line_m1) * (256888.302/(1+105.408 * 2.718281828459045**((-0.183431*line_m1))))
plt.plot(line_m1, polyspace_m1, label = "model 1", color="blue")

# Logistic Model 2
line_m2 = np.linspace(0, 64, 100) 
polyspace_m2 = ((1.02016/1.02493)**(0.84*line_m1)) * (256888.302/(1+105.408 * 2.718281828459045**((-0.15408204*line_m1))))
plt.plot(line_m2, polyspace_m2, label = "model 2", color = "red")

ax.legend()
plt.title("Logistic Model with Curb Effect")
```
